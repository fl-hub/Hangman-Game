```python
def hangman_game():
  def main():

    list_of_words = list(english_words_alpha_set)
    print('==============================================================')
    print('=================== HANGMAN GAME =============================')

    # Download the words dataset (if not already downloaded)
    #nltk.download('words')

    # Settings
    print("Choose the game difficulty:")
    difficulty_levels = {"Easy": 5, "Normal": 8, "Hard": float('inf')}

    print("Available difficulty levels:")
    for difficulty in difficulty_levels.keys():
      print(difficulty)
    dif_selection = input()

    # Get the difficulty level based on user input
    selected_difficulty = difficulty_levels.get(dif_selection.capitalize(), 5)

    # Word Generation
    def generate_random_word(selected_difficulty):
        # Get the list of English words
        word_list = list_of_words

        # Filter out words based on the selected difficulty level
        if selected_difficulty == float('inf'):
            valid_words = word_list
        else:
            valid_words = [word for word in word_list if len(word) <= selected_difficulty]

        # Choose a random word from the filtered list
        random_word = random.choice(valid_words)

        return random_word

    # Game
    random_word = generate_random_word(selected_difficulty)
    word = random_word.upper()
    #used_letters = []

    hangman={0:'''
            ____________
             |''',
            1:'''
            ____________
             |
             O''',
            2:'''
            ____________
             |
             O
            /''',
            3:'''
            ____________
             |
             O
            / \\''',
            4:'''
            ____________
             |
             O
            / \\
             |''',
            5:'''
            ____________
             |
             O
            / \\
             |
            /''',
            6:'''
            ____________
             |
             O
            / \\
             |
            / \\ '''}
    hidden_word = '-'*len(random_word)
    print(f'Hidden word is {hidden_word}')
    mistake_count = 7

    print('==============================================================')

    while mistake_count > 0 and hidden_word.count("-") != 0:
      index_count = 0
      print('Pick a letter:')

      while True:
        picked_letter = input().capitalize()
        #print("Used letters:"+", ".join(used_letters))

        # Check if the input is a single character and not a number
        if len(picked_letter) == 1 and not picked_letter.isdigit():
          #used_letters.append(input)
          break
        else:
            print("Invalid input. Please enter a single character, not a number.")


      for letter in word:
        if letter == picked_letter:
          hidden_word = hidden_word[:index_count] + picked_letter + hidden_word[index_count+1:]

        index_count += 1

      if word.count(picked_letter) > 0:
        print(f'CORRECT! The word contains the {picked_letter}')
        print(hidden_word)
      else:
        mistake_count -= 1
        print(f'WRONG! Number left of mistakes: {mistake_count}')
        print(hangman[6 - mistake_count])
        print(hidden_word)
      print('==============================================================')

    if mistake_count == 0:
      print('======================== GAME OVER!  =========================')
      print('===================== You were hanged! =======================')
      print(f"The word was: {random_word}")
    else:
      print('========================= VICTORY!  ==========================')

  # Function to ask the user if they want to restart
  def ask_to_restart():
    return input("Do you want to restart? (yes/no): ").lower() == 'yes'

  # Initial run
  main()

  # Ask if the user wants to restart
  while ask_to_restart():
    main()
```

```python
hangman_game()
```
